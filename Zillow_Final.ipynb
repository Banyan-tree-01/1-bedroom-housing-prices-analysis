```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
#Zillow Home Value Index (ZHVI) 1 Bedroom count (raw ZHVI)
df_county_ts=pd.read_csv("D:/Python Project Files/(Zillow_project_data) County_zhvi_bdrmcnt_1_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")
df_county_ts
```

```python
df_county=df_county_ts.head(n=100)
```

```python
df_county
```

```python
df_county1=df_county.drop(['RegionID','RegionType','StateName','State','Metro','StateCodeFIPS','MunicipalCodeFIPS','SizeRank'],axis=1)
```

```python
df_county1=df_county1.transpose()
```

```python
df_county1
```

```python
df_county1.columns=df_county1.iloc[0].values
```

```python
df_county1=df_county1.drop(df_county1.index[0])
```

```python
df_county1
```

```python
df_county1.index
```

```python
df_county1.index=pd.to_datetime(df_county1.index)
```

```python
df_county1=df_county1.bfill().ffill()
```

```python
df_county1
```

```python
df_county1.isnull().sum()
```

```python
df_county1.groupby(df_county1.index.year)['Los Angeles County'].mean().plot(kind='bar',figsize=(10,6))
plt.suptitle("Typical value of 1 bedroom apartment in Los Angeles County")
plt.ylabel("Value")
plt.xlabel("Year")
plt.show()
```

Los Angeles County shows a clear long-term appreciation trend in 1-bedroom home values. Prices were relatively flat in the early 2000s, surged during the pre-2008 housing boom, 
dipped after the financial crisis, and then entered a prolonged bull run from around 2012, nearly quadrupling from the post-crisis lows to 2025. The 2020–2022 period shows an 
especially sharp spike, likely driven by pandemic-era low interest rates and housing demand.

```python
county_mean = df_county1.mean(axis=1)

# Create a new DataFrame with the date index and mean values
mean_df = pd.DataFrame({'mean_value': county_mean}, index=df_county1.index)

# Extract year from the index for grouping
mean_df['year'] = mean_df.index.year

# Group by year and calculate the mean for each year
yearly_mean = mean_df.groupby('year')['mean_value'].mean()

# Plot the yearly means
yearly_mean.plot(kind="bar", figsize=(10, 6))
plt.suptitle("County Average over Time")
plt.xlabel("Year")
plt.ylabel("Value")
plt.tight_layout()
```

The cross-county average mirrors the national housing narrative: gradual growth through the early 2000s, a bubble peak around 2006–2007, a sharp correction post-2008, and a strong 
recovery beginning around 2012. The post-2020 surge is the steepest in the dataset's history, reflecting unprecedented monetary policy and shifts in housing demand.

```python
df_county1['quarter']=pd.PeriodIndex(df_county1.index,freq='Q')
```

```python
df_county1
```

```python
quarter_mean=df_county1.groupby(by=df_county1['quarter']).mean()
```

```python
quarter_mean
```

```python
ax3=quarter_mean['Los Angeles County'].plot(figsize=(10,6))
plt.xlabel('quarter')
plt.ylabel('value')
plt.title("time series")
plt.tight_layout()
```

This line chart traces the quarterly average 1-bedroom home value in Los Angeles County from 2000 to 2025, making the full housing cycle visible in one view. Three distinct phases stand out: the boom-and-bust of the mid-2000s subprime era, the slow post-2008 recovery through the 2010s, and the steep post-2020 acceleration driven by pandemic-era demand and low mortgage rates. The recent plateau and slight softening toward 2024–2025 may reflect the impact of rising interest rates cooling an overheated market. LA County's trajectory is a microcosm of broader US coastal housing dynamics.

```python
fig, ax = plt.subplots(figsize=(12,8))
top5 = ['Los Angeles County', 'Cook County', 'San Diego County', 'Maricopa County', 'Harris County']
for county in top5:
    ax.plot(quarter_mean.index, quarter_mean[county], label=county, linewidth=2)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.legend()
ax.set_title('Top 5 US Counties: Quarterly Home Price Index (2000-2025)')
plt.tight_layout()
plt.show()
```

Plotting the top 5 counties quarterly reveals diverging trajectories. LA and San Diego follow similar coastal California patterns with high absolute values. Cook County (Chicago) and Harris County (Houston) grew modestly and more steadily. Maricopa County (Phoenix) saw explosive growth 
post-2020 before showing signs of cooling by 2024–2025 — consistent with its reputation as a pandemic boomtown that overheated.
